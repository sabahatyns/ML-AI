Categorical Variables and One Hot Encoding


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./files/homeprices_03.csv")
df
```

```python
dummies = pd.get_dummies(df["town"], dtype= int)
dummies
```

```python
merged = pd.concat([df, dummies], axis = 'columns') # join both
merged
```

```python
final = merged.drop(['town' , 'west windsor'], axis = 'columns')
final
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression() # model created
```

```python
# differentiate x and y 
# in this data price is y as its dependent variable

x = final.drop('price' , axis= 'columns')
x
```

```python
y = final.price
y
```

```python
# train model
model.fit(x,y)
```

```python
model.predict([[2800, 0, 1]]) # area, monore town and robinsville
```

```python
model.predict([[3400,0,0]])
```

```python
# check how accurate model is
model.score(x, y)
```

### USING ONE HOT ENCODER

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town= le.fit_transform(dfle.town)
dfle
```

```python
x = df[['town', 'area']].values
x
```

```python
y = dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
x = x[:,1:]
x
```

```python
model.fit(x, y)
```

```python
model.predict([[0,1,3400]])
```

```python
model.predict([[1,0,2800]])
```
