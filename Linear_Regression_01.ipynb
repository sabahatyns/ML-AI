```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel("area(sqr ft)")
plt.ylabel("price(US$)")
plt.scatter(df.area,df.price,color='red', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel("area(sqr ft)", fontsize = 20)
plt.ylabel("price(US$)" , fontsize = 20)
plt.scatter(df.area,df.price,color='red', marker='*')
plt.plot(df.area,reg.predict(df[["area"]]),color='blue')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

# Load dataset
df = pd.read_csv("canada_per_capita_income.csv")   # adjust path if needed

# Inspect data
print(df.head())

# Train regression model
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df["per capita income (US$)"])

# Predict income for 2020
prediction = reg.predict(pd.DataFrame({"year": [2020]}))
print(f"Predicted per capita income in 2020: {prediction[0]}")

# Plot data + regression line
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df["year"], df["per capita income (US$)"], color="red", marker="*")
plt.plot(df["year"], reg.predict(df[["year"]]), color="blue")
plt.show()
```
